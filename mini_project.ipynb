SPLITTING DATA IN TEST AND TRAIN FOR FARE_AMOUNT AND TOTAL_AMOUNT

```python
x=df2.drop(columns=["fare_amount","total_amount"])
z=df2["total_amount"]
y=df2["fare_amount"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

AAPLYING MODEL FOR FARE AMOUNT

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.score(x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=42)
cross_val_score(LinearRegression(),x,y,cv=cv)
```

APPLYING MODEL FOR TOTAL AMOUNT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,z_train,z_test=train_test_split(x,z,test_size=.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,z_train)
z_pred=model.predict(x_test)
print(model.score(x_test,z_test))
```

METRICES

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

print(f'R-squared: {r2}')
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the RMSE value
print(f'Root Mean Squared Error: {rmse}')
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
```
